```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("DATA/00-puppy.jpg")
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# If you want to convert to HSV then,
img = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(img)
```

```python
# If you want to convert to HLS then,
img = cv2.cvtColor(img,cv2.COLOR_HSV2RGB)    # converting HSV to RGB

img = cv2.cvtColor(img,cv2.COLOR_RGB2HLS)

plt.imshow(img)
```
